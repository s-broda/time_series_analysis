```julia
using Plots
```

```julia
tarch(u, sig=1.09) = 4.180995387207403e-4+0.08348091252521408u^2+0.2573764220576841*(u^2*(u<0))+0.755583899346605*sig^2
```

```julia
garch(u, sig=1.09) = 4.497635631082216e-04 + 0.2237502674257507*u^2+0.7413093493257758*sig^2
```

```julia
plot(garch, -3, 3, label="GARCH", xlabel="\$u_t\$ in percent", ylabel="\$\\sigma^2_{t+1}\$")
plot!(tarch, -3, 3, label="TARCH")
```
